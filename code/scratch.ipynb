# path class

```python
from os import path
from collections import OrderedDict
```

```python
class Path(OrderedDict):

    def __init__(self, dirname=None, ext=None, **kwargs) -> None:
        super().__init__(**kwargs)
        self.ext = ext
        self.dirname = dirname
        self.basename = self.stitch_(**self)

    @property
    def fpath(self) -> str:
        self.basename = self.stitch_(**self)
        filename = self.basename
        if (ext := self.ext) is not None:
            if not ext.startswith('.'):
                ext = '.' + ext
            filename += ext
        
        if self.dirname is not None:
            filename = path.join(self.dirname, filename)
        return filename


    def update(self, dirname=None, ext=None, **kwargs):
        super().update(**kwargs)
        if dirname is not None:
            self.dirname = dirname
        if ext is not None:
            self.ext = ext
        return self


    def __repr__(self) -> str:
        return self.fpath

    def __str__(self) -> str:
        return self.fpath

    @staticmethod
    def frompath(filename: str, castint=True):
        dirname = path.dirname(filename)
        basename = path.basename(filename)
        string, ext = path.splitext(basename)
        parts = {}
        for part in string.split('_'):
            kv = part.split('-', 1)
            value = None
            if len(kv) > 1:
                value = kv[1]
                if castint and value.isnumeric():
                    value = int(value)
            parts[kv[0]] = value
        return Path(**parts, ext=ext, dirname=dirname)

    @staticmethod
    def stitch_(**kwargs: dict) -> str:
        string = '_'.join(f'{key}-{value}' for key, value in kwargs.items())
        return string
```
